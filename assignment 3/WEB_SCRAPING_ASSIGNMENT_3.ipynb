## 5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.

```python
driver = webdriver.Chrome()
driver.get('https://www.google.com/maps/@51.4623945,-0.999798,15z?entry=ttu')
```

```python
time.sleep(5)
driver.find_element(By.XPATH, '/html/body/c-wiz/div/div/div/div[2]/div[1]/div[3]/div[1]/div[2]/form[1]/div/div/button/span').click()
```

```python
designation = driver.find_element(By.XPATH, '/html/body/div[3]/div[8]/div[3]/div[1]/div[1]/div/div[2]/form/input')
designation.send_keys("London")
```

```python
button_search = driver.find_element(By.XPATH, '/html/body/div[3]/div[8]/div[3]/div[1]/div[1]/div/div[2]/div[1]/button')
button_search.click()
```

```python
import re

url_string = driver.current_url

# Regular expression pattern to match the coordinates
pattern = r'@([-0-9.]+),([-0-9.]+)'

matches = re.search(pattern, url_string)
if matches:
    latitude = matches.group(1)
    longitude = matches.group(2)
    print("Latitude:", latitude)
    print("Longitude:", longitude)
else:
    print("Coordinates not found in the URL.")

# Close the Selenium driver
driver.quit()
```
