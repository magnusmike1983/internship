```python
from bs4 import BeautifulSoup as bs
import requests

url = "https://presidentofindia.nic.in/former-presidents.htm"
page = requests.get(url)
page

soup = bs(page.content, "html.parser")
soup

PresidentNames_TermOfOffice = []

data = soup.find_all("div", class_="presidentListing")

for i in data:
    PresidentNames_TermOfOffice.append(i.text.strip())
PresidentNames_TermOfOffice
```

```python
P = PresidentNames_TermOfOffice[0].split('\n')
P
```

```python
len(PresidentNames_TermOfOffice)
```

```python
data = []
for x in range (0,14):
    P = PresidentNames_TermOfOffice[x].split('\n')
    data.append(P)
data
```

```python
headers = ["Name", "Term_of_office", "website"]
```

```python
df = pd.DataFrame(data, columns=headers, dtype=str)
df
```

```python
x = df['Term_of_office']
x
```

```python
x= x.str.replace('Term of Office:', '').str.strip()
x
```
